```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load,)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# District Summary

- Calculate the total number of schools
- Calculate the total number of students
- Calculate the total budget
- Calculate the average math score 
- Calculate the average reading score
- Calculate the percentage of students with a passing math score (70 or greater)
- Calculate the percentage of students with a passing reading score (70 or greater)
- Calculate the percentage of students who passed math and reading (% Overall Passing)
- Create a dataframe to hold the above results
- Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].unique())

# Calculate the total number of students
total_students = school_data_complete["student_name"].count()

#Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average math score
average_maths_score = round(school_data_complete["math_score"].mean(), 6)

# Calculate the average reading score
average_reading_score = round(school_data_complete["reading_score"].mean(), 5)

# Calculate the percentage of students with a passing math score (70 or greater)
student_greater70_math = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
percent_math_score70 = round((student_greater70_math["math_score"].count() / total_students) * 100, 6)

# Calculate the percentage of students with a passing reading score (70 or greater)
student_greater70_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70, : ]
percent_read_score70 = round((student_greater70_reading["reading_score"].count() / total_students) * 100, 6)

# Calculate the percentage of students who passed math and reading (% Overall Passing)
student_overall = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70), : ]
percent_overall = round((student_overall["reading_score"].count() / total_students) * 100, 6)

# Calculate the percentage of students who passed math and reading (% Overall Passing)
district_summary_df = pd.DataFrame({
    "Total Schools" : [total_schools],
    "Total Students" : [total_students], 
    "Total Budget" : [total_budget],
    "Average Math Score" : [average_maths_score], 
    "Average Reading Score" : [average_reading_score], 
    "% Passing Math" : [percent_math_score70],
    "% Passing reading" : [percent_read_score70], 
    "% Overall Passing" : [percent_overall]
})

# Format the above data frame
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,.0f}".format)

district_summary_df

```

# School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create a copy of school_data and student_data
school_data_df = school_data.copy()
student_data_df =student_data.copy()

# Insert a new column 'Per Student Budget' in the 'school_data_copy' DataFrame and calculate the budget per student
school_data_df["Per Student Budget"] = school_data_df["budget"] / school_data_df["size"]

# Group the 'student_data_df' DataFrame, calculate the average of math and reading score and reset the index
average_math_score = student_data_df.groupby(["school_name"])["math_score", "reading_score"].mean()
average_math_score.reset_index()

# Merge the above DataFrame to 'school_data_df' DataFrame on key "school_name"
school_data_df = school_data_df.merge(average_math_score, on = "school_name", how = "outer")

# Rename the column headers in the main DataFrame i.e 'school_data_df'
school_data_df = school_data_df.rename(columns = {"math_score" : "Average Math Score",
                                                 "reading_score" : "Average Reading Score"
                                                 })

# Locate the students whose math's score is greater than or equal to 70
passing_math70 = student_data_df.loc[student_data_df["math_score"] >= 70, :]

# Group the 'passing_math70' DataFrame by 'school_name', count the number of students who scored >= 70 and reset the index
total_passing70_math = passing_math70.groupby(["school_name"])["math_score"].count() 
total_passing70_math.reset_index()

# Merge the above DataFrame to 'school_data_df' DataFrame on key "school_name"
school_data_df = school_data_df.merge(total_passing70_math, on = "school_name", how = "outer")

# Calculate the '% Passing Math' and store the data in 'math_score' column
school_data_df["math_score"] = (school_data_df["math_score"] / school_data_df["size"]) * 100

# Locate the students whose reading's score is greater than or equal to 70
passing_reading70 = student_data_df.loc[student_data_df["reading_score"] >= 70, :]

# Group the 'passing_reading70' DataFrame by 'school_name', count the number of students who scored >= 70 and reset the index
total_passing70_reading = passing_reading70.groupby(["school_name"])["reading_score"].count() 
total_passing70_reading.reset_index()

# Merge the above DataFrame to 'school_data_df' DataFrame on key "school_name"
school_data_df = school_data_df.merge(total_passing70_reading, on = "school_name", how = "outer")

# Calculate the '% Passing Reading' and store the data in 'reading_score' column
school_data_df["reading_score"] = (school_data_df["reading_score"] / school_data_df["size"]) * 100

# Rename the column headers in the main DataFrame i.e 'school_data_df'
school_data_df = school_data_df.rename(columns = {"math_score" : "% Passing Math",
                                                 "reading_score" : "% Passing Reading"
                                                 })

# Locate the students that passed both math and reading with score >= 70
passing_math70_reading70 = student_data_df.loc[(student_data_df["math_score"] >= 70) & (student_data_df["reading_score"] >= 70), :]

# Group the 'passing_math70_reading70' DataFrame by 'school_name', count the number of students who scored >= 70 both in math & reading and reset the index
# By counting the column, either 'math_score' or 'reading_score', we will get the number of students who passed in both math and reading
total_math70_reading70 = passing_math70_reading70.groupby(["school_name"])["math_score"].count()
total_math70_reading70.reset_index()

# # Merge the above DataFrame to 'school_data_df' DataFrame on key "school_name"
school_data_df = school_data_df.merge(total_math70_reading70, on = "school_name", how = "outer")

# Calculate the '% Overall Passing' and store the data in 'math_score' column
school_data_df["math_score"] = (school_data_df["math_score"] / school_data_df["size"]) * 100

# Rename the column headers in the main DataFrame i.e 'school_data_df'
school_data_df = school_data_df.rename(columns = {"school_name" : "School Name",
                                                  "type" : "School Type",
                                                  "size" : "Total Students",
                                                  "budget" : "Total School Budget",
                                                  "math_score" : "% Overall Passing",
                                                 })


# Delete 'School ID' column in the DataFrame
del school_data_df["School ID"]

# Sort the DataFrame in ascending order based on 'School Name' column
school_data_df = school_data_df.sort_values("School Name", ascending = True)

# Format the above data frame to add '$' sign and ','
school_data_df["Total School Budget"] = school_data_df["Total School Budget"].map("${:,.2f}".format)
school_data_df["Per Student Budget"] = school_data_df["Per Student Budget"].map("${:.2f}".format)

# school_data_df = school_data_df.reset_index(level = None)
school_data_df
```
